```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# LOAD DATA
df1 = pd.read_csv("unemployment in india.csv")
df2 = pd.read_csv("unemployment_rate_upto_11_2020.csv")

# REMOVE EXTRA SPACES IN COLUMN NAMES (IMPORTANT FIX)
df1.columns = df1.columns.str.strip()
df2.columns = df2.columns.str.strip()

# CHECK COLUMNS
print("DF1 Columns:", df1.columns)
print("DF2 Columns:", df2.columns)

# CLEAN DATA
df1 = df1.dropna()
df2 = df2.dropna()

# FIND DATE COLUMN AUTOMATICALLY (SAFE METHOD)
date_col = None
for col in df1.columns:
    if "date" in col.lower():
        date_col = col

if date_col:
    df1[date_col] = pd.to_datetime(df1[date_col], errors='coerce')
    df1 = df1.dropna(subset=[date_col])
    df1 = df1.sort_values(date_col)

# FIND UNEMPLOYMENT COLUMN AUTOMATICALLY
rate_col = None
for col in df1.columns:
    if "unemployment" in col.lower():
        rate_col = col

print("Detected Rate Column:", rate_col)

# PLOT TREND
plt.figure(figsize=(12,6))
plt.plot(df1[date_col], df1[rate_col])
plt.title("Unemployment Trend in India")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate")
plt.show()

# COVID ANALYSIS
covid_start = "2020-03-01"

before = df1[df1[date_col] < covid_start]
during = df1[df1[date_col] >= covid_start]

print("\nCOVID ANALYSIS")
print("Before COVID:", before[rate_col].mean())
print("During COVID:", during[rate_col].mean())

# MONTHLY TREND
df1['Month'] = df1[date_col].dt.month

monthly = df1.groupby('Month')[rate_col].mean()

plt.figure(figsize=(10,5))
monthly.plot(kind='bar')
plt.title("Monthly Unemployment Trend")
plt.show()

# HEATMAP
plt.figure(figsize=(8,5))
sns.heatmap(df1.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()

# FINAL INSIGHTS
print("\nFINAL INSIGHTS")
print("Average Unemployment:", df1[rate_col].mean())
print("COVID Impact Increase:", during[rate_col].mean() - before[rate_col].mean())
```
